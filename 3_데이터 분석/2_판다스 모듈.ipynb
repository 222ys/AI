# **18. 데이터프레임 병합**

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol') # 기존 데이터프레임
df2 = pd.read_csv('https://bit.ly/ds-korean-idol-2') # 추가할 데이터프레임
```

```python
df_copy = df.copy() # 원본 보존
```

```python
# row에 데이터를 합칠 경우에는 pd.concat()를 사용하여 데이터프레임을 list로 합침
# row 기준으로 합칠 경우 sort=False 옵션을 주어 순서가 유지되도록 함
pd.concat([df, df_copy], sort=False)
```

```python
df_concat = pd.concat([df, df_copy], sort=False) # 결과를 변수에 저장
df_concat.reset_index() # 병합 후 인덱스 정렬
```

```python
df_concat.reset_index(drop=True) # 새 인덱스 생성 및 기존 인덱스 제거
```

```python
# column을 기준으로 합칠 경우 axis=1 옵션을 설정
pd.concat([df, df2], axis=1)
```

```python
df3 = df2.drop([3, 5]) # 데이터 수가 안 맞게 해보기
pd.concat([df, df3], axis=1) # 데이터 수가 안 맞아도 index 맞춰서 NaN 채움
```

* concat : row나 column 기준으로 단순하게 이어붙임
* merge : 특정 고유한 키(unique id)값을 기준으로 붙임

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
df2 = pd.read_csv('https://bit.ly/ds-korean-idol-2')
```

```python
df_right = df2.drop([1, 3, 5, 7, 9], axis=0) # 1, 3, 5, 7, 9번 행 제거
df_right = df_right.reset_index(drop=True) # 새 인덱스 생성 및 기존 인덱스 제거
df_right
```

```python
pd.concat([df, df_right], axis=1) # 데이터 수도 안맞고 인덱스까지 안맞으니 전부 틀어짐
```

```python
pd.merge(df, df_right, on='이름', how='left') # 이름이라는 고유한 키를 기준으로, left를 기준으로 병합
```

```python
pd.merge(df, df_right, on='이름', how='right') # 이름이라는 고유한 키를 기준으로, right를 기준으로 병합
```

```python
# inner 방식은 두 데이터프레임에 모두 키 값이 존재하는 경우만 병합
pd.merge(df, df_right, on='이름', how='inner')
```

```python
# outer 방식은 왼쪽, 오른쪽 데이터를 모두 보여주는 병합
pd.merge(df, df_right, on='이름', how='outer') # 여기서는 left병합한 결과와 같음. left가 right의 데이터를 모두 포함하면서 데이터 수도 더 많아서.
```

```python
df_right.columns=['성함', '연봉', '가족수'] # 고유 키였던 '이름'을 '성함'으로 바꿔봄
df_right
# pd.merge(df, df_right, on='이름', how='outer') : 키이름이 다르므로 에러 뜸
```

```python
pd.merge(df, df_right, left_on='이름', right_on='성함', how='outer')
# 왼쪽의 이름과 오른쪽의 성함을 각각 키로 설정해 병합. 단, 중복되는 성함 열이 삭제되지는 못함
```

# **25. 원 핫 인코딩(One-Hot-Encoding)**

* 원 핫 인코딩은 한 개의 요소는 True, 나머지 요소는 False로 만들어주는 기법
* 각각의 데이터가 관계성을 갖지 않도록 서로 카테고리컬하게 기계에게 알려주는 것
* 데이터를 개수만큼 뿌려주고 개수에 맞는 데이터에만 True, 나머지에는 False를 넣어주는 기법
* 데이터 전처리 기법 중 하나

```python
blood_map = {'A':0, 'B':1, 'AB':2, 'O':3} # 딕셔너리 형식
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map) # map은 딕셔너리 사용, replace 형식
df
```

```python
df['혈액형_code'].value_counts()
```

* df['혈액형_code']를 머신러닝 알고리즘에 넣어 데이터를 예측하라고 지시한다면 컴퓨터는 값들간의 관계를 스스로 형성하게 됨
* B형(1) + AB형(2) = O형(3) 과 같은 잘못된 관계를 맺어버릴 수 있음
* 따라서 별도의 column을 형성해주고, 1개의 column에는 True 나머지는 False를 넣어줌으로써 A, B, AB, O의 관계는 독립적이라는 카테고리로 표혀해주는 방식

```python
pd.get_dummies(df['혈액형_code']) # 원핫인코딩 코드
```
