### **1.2 워드클라우드에 마스크 씌우기**

```python
text = open('./alice.txt').read()
print(text)
```

```python
from PIL import Image # 이미지에서 마스킹 데이터를 출력
from wordcloud import STOPWORDS # 불용어(워드 클라우드로 작성할 단어에서 제외할 단어)
```

### **1.3 워드클라우드에 회색 입히기**

```python
import random
```

```python
# word: 표시할 단어
# font_size: 폰트 크기
# position: 표시할 단어의 위치
# orientation: 표시할 단어의 방향
# hwargs: 단어에 적용할 글꼴의 경로
# hsl(색상, 채도, 밝기)

def gray_color_func(word, font_size, position, orientation=None, **hwargs):
  return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100) # %가 두 개면 진짜 %를 의미함. 따라서 0, 0%, 랜덤%가 됨
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

### **1.4 연습문제-회색입히기**

```python
# 워크클라우드 작성 순서
# 1. a_new_hope.txt 불러오기
# 2. stormtrooper_mask.png 마스크 불러오기
# 3. STOPWORDS 객체 생성
# 4. int, ext 단어 추가
# 5. 배경은 흰색, 단어는 1000개, mask와 stopwords 적용
# 6. 회색으로 덮어쓰기
```

```python
text = open('./a_new_hope.txt').read()

import numpy as np
from PIL import Image
stormtrooper_mask = np.array(Image.open('./stormtrooper_mask.png'))

from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')

wc = WordCloud(background_color = 'white', max_words=1000, stopwords=stopwords, mask=stormtrooper_mask)
wordcloud=wc.generate(text)

def gray_color_func(word, font_size, position, orientation=None, **hwargs):
  return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

### **1.6 연습문제-형태소분석과 워드클라우드 융합**

```python
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t=Twitter()
```

```python
ko_law_text = kolaw.open('constitution.txt').read()
tokens_ko = t.nouns(ko_law_text)
```

```python
tokens_ko
```

```python
# 글자수가 1개인 불용어 제거
tokens_ko = [i for i in tokens_ko if len(i) > 1]
tokens_ko
```

```python
# Text(): 문서를 편리하게 탐색할 수 있는 다양한 기능 제공
import nltk
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
data = ko.vocab().most_common(500)
tmp_data = dict(data)
tmp_data
```
