# **2. 행렬 연산**

### **2-1. Matrix**
* 행(row) + 열(column)
* 데이터분석 및 기계학습에서는 행렬을 제외하고 연산할 수 없음

### **2-2. 덧셈 연산**
* shape가 같아야 함
* 같은 position끼리 연산

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
a + b
```

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[4, 5],
              [2, 3]])
```

```python
# a + b : ValueError
```

### **2-3. 뺄셈 연산**
* shape가 같아야 함
* 같은 position끼리 연산

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
a - b
```

### **2-4. 곱셈 연산**
* shape가 같아야 함
* 같은 position끼리 연산 

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
a * b
```

### **2-5. dot product**
* 맞닿는 shape가 같아야 함

```python
a = np.array([[1, 2, 3],
              [1, 2, 3],
              [2, 3, 4]])
b = np.array([[1, 2],
             [3, 4],
             [5, 6]])
```

```python
a.shape, b.shape
```

```python
np.dot(a, b) # (앞 행 * 뒤 열)들의 합
```

```python
# (1*1 + 2*3 + 3*5), (1*2 + 2*4 + 3*6)
# (1*1 + 2*3 + 3*5), (1*2 + 2*4 + 3*6)
# (2*1 + 3*3 + 4*5), (2*2 + 3*4 + 4*6)
```

### **2-7. 정렬(sort)**

```python
arr = np.array([1, 10, 5, 8, 2, 4, 3, 6, 8, 7, 9])
```

```python
arr
```

```python
np.sort(arr) # 기본적으로 오름차순 정렬
```

```python
np.sort(arr)[::-1] # 내림차순
```

```python
arr # 파이썬 기본 정렬과 달리 값이 유지되지 않기 때뮨에 변수에 저장해야 함
```

```python
result = np.sort(arr)
```

```python
result
```

```python
arr
```

```python
arr2d = np.array([[11, 10, 12, 9],
                 [3, 1, 4, 2],
                 [5, 6, 7, 8]])
```

```python
arr2d.shape
```

```python
# 행정렬 : 각 행의 맨 앞 숫자를 기준으로 행 단위로 정렬된다
np.sort(arr2d, axis=0)
```

```python
# 열정렬 : 행 별로 그 안의 열을 비교해 정렬된다
np.sort(arr2d, axis=1)
```

```python
# 마지막 방향 (여기서는 2차원이므로 열정렬)
np.sort(arr2d, axis=-1)
```
