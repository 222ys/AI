* 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조
* 줄을 서는 행위와 유사
* FIFO(First-in, First-out)

### **1-1. 큐의 용어**
* Enqueue : 큐에 데이터를 넣는 기능
* Dequeue : 큐에서 데이터를 꺼내는 기능
* [Visualgo](https://visualgo.net/en/list)


### **1-2. 큐의 사용**
* 멀티태스킹을 위한 프로세스 스케쥴링 방식을 구현(운영체제)
* 푸시메세지

**1-3. 파이썬 queue 라이브러리를 활용하여 queue 자료구조 사용**
* Queue() : 가장 일반적인 큐 자료구조를 생성
* LifoQueue() : 나중에 입력된 데이터가 먼저 출력되는 구조의 큐 자료구조를 생성(스택구조와 비슷)
* PriorityQueue() : 데이터마다 우선순위를 넣어서 우선순위가 높은순으로 데이터를 출력
* 프로그램을 작성할 때 프로그램에 따라 적합한 자료구조를 사용

```python
# queue로 큐 만들기
import queue

data_queue = queue.Queue()
data_queue.put('Hello')
print(data_queue)
print(data_queue.qsize()) # 데이터의 개수 출력
print(data_queue.get()) # Dequeue 기능
print(data_queue.qsize())
```

```python
data_queue.put(10)
data_queue.put(True)
print(data_queue.qsize())
print(data_queue.get())
```

```python
data_queue.put(20)

for i in range(data_queue.qsize()):
  item = data_queue.get()
  print(item)
```

```python
print(data_queue.qsize())
```

```python
# LifoQueue()로 큐 만들기
data_queue = queue.LifoQueue()
data_queue.put('Hello')
data_queue.put(10)
data_queue.put(True)
print(data_queue.qsize())
print(data_queue.get())
```

```python
# PriorityQueue()로 큐 만들기
data_queue = queue.PriorityQueue()
data_queue.put((10, '김사과'))
data_queue.put((5, '반하나'))
data_queue.put((7, '오렌지'))
data_queue.put((8, '이메론'))
data_queue.put((2, '박체리'))
print(data_queue.qsize())
print(data_queue.get()) # 키값이 작을수록 우선순위가 높음
print(data_queue.get())
```

### **문제.**

리스트 변수로 큐를 다루는 enqueue, dequeue 기능을 구현해보자.

```python
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  print(queue_list[0])
  queue_list.pop(0)
```

```python
enqueue(10)
enqueue(20)
enqueue(30)
dequeue()
```

```python
# 정답
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data = queue_list[0]
  del queue_list[0]
  return data
```

```python
print(len(queue_list))
```

```python
for index in range(10):
  enqueue(index)
```

```python
print(len(queue_list))
queue_list
```
