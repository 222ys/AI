## **1. 더블 링크드리스트(Double Linked List)**

* 양방향으로 연결되어 있어서 노드 탐색이 양쪽으로 모두 가능

```python
class Node:
  def __init__(self, data, prev=None, next=None):
    self.prev = prev
    self.data = data
    self.next = next
```

```python
class NodeMgmt:
  def __init__(self, data):
    self.head = Node(data)
    self.tail = self.head

  def insert(self, data):
    if self.head == None:
      self.head = Node(data)
      self.tail = self # Node(data) 또 하면 안됨
    else:
      node = self.head
      while node.next:
        node = node.next
      new = Node(data)
      node.next = new
      new.prev = node
      self.tail = new

  def node_print(self):
    node = self.head
    while node:
      print(node.data)
      node = node.next

  # 앞에서부터 검색하여, 전달한 data 값이 존재하는 경우 해당 값을 출력하고, 존재하지 않는 경우 False를 출력
  def search_from_head(self, data):
    if self.head == None:
      return False
    node = self.head
    while node:
      if node.data == data:
        return node.data
      else:
        node = node.next
    return False

  # 뒤에서부터 검색하여, 전달한 data 값이 존재하는 경우 해당 값을 출력하고, 존재하지 않는 경우 False를 출력
  def search_from_tail(self, data):
    if self.tail == None:
      return False
    node = self.tail
    while node:
      if node.data == data:
        return node.data
      else:
        node = node.prev
    return False

  def insert_before(self, data, after_data): # insert_before(2.5, 3)
    if self.head == None:
      self.head = Node(data)
      return True
    else:
      node = self.head
      while node.data != after_data:
        node = node.next
        if node == None:
          print('해당 숫자의 노드가 존재하지 않습니다')
      new = Node(data)
      before_new = node.prev
      before_new.next = new
      new.prev = before_new
      new.next = node
      return True
```

```python
dblist = NodeMgmt(0)
for data in range(1, 11):
  dblist.insert(data)
```

```python
dblist.node_print()
```

```python
print(dblist.head)
print(dblist.tail)
```

```python
search_result = dblist.search_from_head(7)
print(search_result)
```

```python
search_result = dblist.search_from_head(17)
print(search_result)
```

```python
search_result = dblist.search_from_tail(7)
print(search_result)
```

```python
search_result = dblist.search_from_tail(17)
print(search_result)
```
