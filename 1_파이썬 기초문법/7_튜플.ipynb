# **1. 튜플(Tuple)**

```python
tu1 = ()
print(tu1)

# tu1[0] = 10 # TypeError: 'tuple' object does not support item assignment

tu2 = (1)
print(tu2)

tu3 = 1, 2, 3
print(tu3)

tu4 = ('apple', 'banana', ('😀', '😁'))
print(tu4)
```

```python
print(type(tu2)) # Q. 왜 tuple이 아니라 integer로 인식하는지? -> tu2 = (1,) 처럼 저장해야 함
print(type(tu3))
print(type(tu4))
```

# **2. 튜플의 수정 및 삭제**

```python
tu5 = (1, 2, 'apple', 'banana')
print(tu5)

# tu5[3] = 'orange' # 튜플의 요소는 수정 불가
# del tu5[3] # 튜플의 요소는 삭제 불가

tu5 = 10 # 아예 주소를 바꾸는 것이므로 가능
print(tu5)

del tu5 # 아예 튜플을 삭제하는 것이므로 가능
# print(tu5) # NameError: name 'tu5' is not defined
```

# **3. 튜플의 인덱싱과 슬라이싱 연산**

```python
tu7 = (1, 2, 'apple', 'banana')

# 인덱싱
print(tu7[0])
print(tu7[-1])

# 슬라이싱
print(tu7[1:])
print(tu7[1:3])

# 덧셈(연결)
tu8 = (30, 40)
print(tu7 + tu8)

# 곱셈(반복)
print(tu8 * 3)

# 요소의 개수
print(len(tu8))
```
