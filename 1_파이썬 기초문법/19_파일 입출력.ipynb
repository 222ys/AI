# **1. 파일 열기**

```
파일변수 = open('텍스트 파일 이름', '파일 열기 모드')
```
* 파일 열기 모드 : w(출력), a(추가출력), r(입력)
* w는 지정된 경로에 파일이 없으면 파일을 만든 후 저장하고, 파일이 있으면 기존의 모든 데이터를 지우고 덮어씀
* a는 지정된 경로에 파일이 없으면 파일을 만든 후 저장하고, 파일이 있으면 기존의 데이터 뒤에 저장
* r은 지정된 경로에 파일이 없으면 에러
* 출력시(w, a) 작업이 완료되면 반드시 close() 메서드를 사용해서 파일을 닫아야 정상적으로 데이터가 저장되고, 입력시(a) 파일을 닫지 않아도 정상적으로 처리됨
* 파일의 종류 구분 : t(text), b(binary)
```
file = open('data.txt', 'r') # 열기모드
file = open('data.txt', 'rt') # 텍스트파일 열기모드
```

````python
file = open('data.txt', 'wt') # t는 안 써도 상관없음
for i in range(10):
  file.write('파일 열기 테스트 : ' + str(i) + '\n')
file.close()
print('data.txt 파일에 쓰기 완료!')
````

* 절대 경로 : 디스크 드라이브의 최상위(root) 디렉토리부터 텍스트 파일이 저장될 디렉토리까지의 경로
  * d:\ryuzy\python\data.txt
* 상대 경로 : 현재 화면에 표시되는 소스 파일이 위치한 디렉토리부터 텍스트 파일이 저장될 디렉토리까지의 경로
  * data.txt(같은 디렉토리)
  * ../data.txt(상위 디렉토리)
  *./test/data.txt(하위 디렉토리)

````python
file = open('./test/data.txt', 'wt')
for i in range(10):
  file.write('하위 파일 열기 테스트 : ' + str(i) + '\n')
file.close()
print('./test/data.txt 파일에 쓰기 완료!')
````

### **문제.**
키보드로 입력하는 문자열을 텍스트 파일에 저장하기  
(단, 'quit'가 입력되면 저장을 끝내고 프로그램을 종료)

````python
while True:
  str = input()
  if str == 'quit':
    file.close()
    break
  file = open('question.txt', 'at')
  file.write(str)
````

````python
# 모범답안
file = open('./test/output.txt', 'w')
while True:
  str = input('>> ')
  if str.lower() == 'quit':
    break
  file.write(str + '\n')
file.close()
print('output.txt 파일 저장완료!')
````

````python
# with 문
# 예외처리 및 close() 메소드를 자동으로 처리함

with open('./test/output.txt', 'w') as file:
  while True:
    str = input('>> ')
    if str.lower() == 'quit':
      break
    file.write(str + '\n')
print('output.txt 파일 저장완료!')
````

````python
# readline(): 텍스트 파일을 한 줄씩 읽어서 처리, 파일이 종료되어 더 이상 읽을 수 없으면 빈 문자열('')을 읽음
file = open('./test/input.txt', 'rt')
lines = []

while True:
  line = file.readline()
  if not line: # 읽어올 줄이 없다면
    break
  if len(line.strip()) != 0:
    print(line, end='')
    lines.append(line.strip())
file.close()
print('input.txt 파일 읽기 완료!')
print(lines)
````

````python
# readlines()
# 전체라인을 모두 읽어서 각 라인단위로 리스트에 저장해주는 메서드
file = open('./test/input.txt', 'rt')
lines = file.readlines()
print(type(lines))
print(lines)
````
