# **1. 상속**

* 기존의 정의해둔 클래스의 기능을 그대로 물려받음
* 기존 클래스에 기능 일부를 추가하거나, 변경하여 새로운 클래스를 정의함
* 코드를 재사용할 수 있게 됨
* 상속을 받고자 하는 대상인 기존 클래스를 부모, parent, super, base 클래스라고 부름
* 상속 받는 새로운 클래스를 자식, child, sub 클래스라고 부름

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def eat(self, food):
    print('{}는 {}를 먹습니다'.format(self.name, food))

  def sleep(self, hour):
    print('{}는 {}시간 동안 잡니다'.format(self.name, hour))
```

```python
animal = Animal('동물', 5)
animal.eat('먹이')
animal.sleep(5)
```

```python
class Dog(Animal): # Animal 클래스를 상속
  pass
```

```python
Lucy = Dog('루시', 12)
Lucy.eat('사료')
Lucy.sleep(10)
```

# **2. 메서드 오버라이딩**

* 부모 클래스의 메서드를 재정의(override)
* 자식 클래스의 인스턴스로 호출시, 재정의된 메서드가 호출됨


```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def eat(self, food):
    print('{}는 {}를 먹습니다'.format(self.name, food))

  def sleep(self, hour):
    print('{}는 {}시간 동안 잡니다'.format(self.name, hour))
```

```python
class Dog(Animal):
  def eat(self, food):
    print('{}는 {}를 아주 맛있게 먹습니다'.format(self.name, food))
  def super_eat(self, food):
    super().eat(food) # 부모객체를 호출해서 입력받은 값을 부모메서드로 넘겨주겠다는 뜻
```

```python
# 기존의 부모클래스로 호출
cat = Animal('고양이', 6)
cat.eat('먹이')
cat.sleep(5)
```

```python
# 자식클래스로 호출, 오버라이딩 
Bori = Dog('보리', 12)
Bori.sleep(8)
Bori.eat('연어') # 오버라이딩
Bori.super_eat('연어') # 오버라이딩 하지않고 부모클래스의 객체 사용
```
