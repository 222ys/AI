# **1. 대입연산자**

* =, +=, -=, /=, %=
* 왼쪽에 저장된 값과 오른쪽의 값을 연산한 뒤 다시 왼쪽에 대입

````python
num = 10

num += 1
print(num)

num -= 1
print(num)

num *= 10
print(num)

num /= 10
print(num)

num %= 2
print(num)
````

# **2. while문**

```
while 조건식:
  조건식이 True인 동안 반복할 문장
  '''
```

````python
i = 1
while i <= 5:
  print('Hello Python')
  i += 1

print('현재 i의 값: {}'.format(i))
````

````python
# 1 ~ 10까지의 합
i = 1
sum = 0
while i <= 10:
  sum += i
  i+= 1
print('1부터 10까지의 합: {}'.format(sum))
````

````python
# 원하는 단을 입력하여 구구단을 출력
dan = int(input('원하는 단을 입력하세요: '))
print('{}단'.format(dan))
i = 1

while i <= 9:
  print('{} * {} = {}'.format(dan, i, dan * i))
  i += 1
````

# **3. for문**

```
for 변수 in range(범위):
  반복할 문장
  '''
```

````python
for i in range(0, 10, 1): # (시작인덱스, 종료인덱스(직전), 증가값)
  print(i, end=' ')
````

````python
for i in range(10): # 시작인덱스는 기본값 0, 증가값은 기본값 1
  print(i, end=' ')
````

````python
for i in range(0, 10): # 증가값은 기본값 1
  print(i, end=' ')
````

````python
# for ( :: ) # 파이썬은 for문에서 무한루프가 안되지만 다른 언어에서는 가능
````

````python
len = range(0, 10, 1)
print(len)

for i in len: # 범위에는 range 말고 다른 것들도 들어갈 수 있다.
  print(i, end=' ')
````

````python
# 1부터 100까지의 총합을 구하는 프로그램
sum = 0

for i in range(1, 101):
  sum += i

print(sum)
````

### **문제 1.**

🎃 🎃 🎃 🎃 🎃  
🎃 🎃 🎃 🎃  
🎃 🎃 🎃  
🎃 🎃  
🎃  

````python
# 내 답안
for i in range(5):
  for j in range(5-i):
    print('🎃', sep=' ',end='')
  print()
````

````python
# 모범답안
for i in range(5):
  for j in range(i, 5):
    print('🎃', end=' ')
  print()
````

### **문제 2.**

````python
# 내 답안
for i in range(5):
  for j in range(i+1):
    print('🎃',sep=' ',end='')
  print()
````

````python
# 모범답안
for i in range(5):
  for j in range(0, i+1):
    print('🎃', end=' ')
  print()
````
