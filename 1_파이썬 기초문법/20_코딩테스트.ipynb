# **1. 수 정렬하기**

https://www.acmicpc.net/problem/2750

N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.

```python
n = int(input())
list = []
for i in range(n):
  list.append(int(input()))
list.sort()
for i in list: print(i)
```

# **2. 음계**

https://www.acmicpc.net/problem/2920

다장조는 c d e f g a b C, 총 8개 음으로 이루어져있다.  
이 문제에서 8개 음은 다음과 같이 숫자로 바꾸어 표현한다.  
c는 1로, d는 2로, ..., C를 8로 바꾼다.

1부터 8까지 차례대로 연주한다면 ascending, 8부터 1까지 차례대로 연주한다면 descending, 둘 다 아니라면 mixed 이다.

연주한 순서가 주어졌을 때, 이것이 ascending인지, descending인지, 아니면 mixed인지 판별하는 프로그램을 작성하시오.

```python
# 내 답안1
code = []
code.extend(map(int,input().split(' ')))
if code == [1, 2, 3, 4, 5, 6, 7, 8]:
  print('ascending')
elif code == [8, 7, 6, 5, 4, 3, 2, 1]:
  print('descending')
else:
  print('mixed')
```

```python
# 내 답안2
num = []
num.extend(map(int, input().split(' ')))

ascending = True
descending = True

for i in range(7):
  if num[i] < num[i+1]:
    descending = False
  elif num[i] > num[i+1]:
    ascending = False
  
if ascending:
  print('ascending')
elif descending:
  print('descending')
else:
  print('mixed')
```

```python
num1=list(map(int, input().split(' ')))

ascending = True
descending = True

for i in range(1, 8):
  if num1[i] > num1[i-1]:
    descending = False
  elif num1[i] < num1[i-1]:
    ascending = False
  
if ascending:
  print('ascending')
elif descending:
  print('descending')
else:
  print('mixed')
```

# **3. 나이순 정렬**

온라인 저지에 가입한 사람들의 나이와 이름이 가입한 순서대로 주어진다. 이때, 회원들을 나이가 증가하는 순으로, 나이가 같으면 먼저 가입한 사람이 앞에 오는 순서로 정렬하는 프로그램을 작성하시오.

https://www.acmicpc.net/problem/10814

```python
n = int(input())

array= []

for _ in range(n):
  data = input().split(' ')
  array.append((int(data[0]), data[1])) # (21, Junkyu)

array = sorted(array, key=lambda x: x[0])

for i in array:
  print(i[0], i[1])
```
