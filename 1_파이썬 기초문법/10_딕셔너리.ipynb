# **1. 딕셔너리(Dictionary)**

* 대응관계를 나타내는 자료형으로 key와 value를 한 쌍으로 갖는 형태
* 하나의 딕셔너리의 키는 중복될 수 없음
* 하나의 딕셔너리의 값은 중복될 수 있음

### **1-1. 딕셔너리 만들기**

```python
dic1 = {}
print(dic1)

dic2 = {'no':1, 'userid':'apple', 'name': '정윤서', 'hp':'010-1111-1111'}
print(dic2)
```

### **1-2. 키로 값 찾기**

```python
dic2 = {'no':1, 'userid':'apple', 'name': '정윤서', 'hp':'010-1111-1111'}
print(dic2['no'])
print(dic2['userid'])
print(dic2['name'])
print(dic2['hp'])
```

### **1-3. 데이터 추가 및 삭제**

```python
dic3 = {1:'apple'}
print(dic3)

dic3[100] = 'orange'
print(dic3)

dic3[50] = 'banana'
print(dic3)

dic3[1] = 'melon' # 키는 중복될 수 없으므로 값이 변경됨
print(dic3)

del dic3[50] # 요소 삭제
print(dic3)
```

### **1-4. 딕셔너리 함수**

```python
dic2 = {'no':1, 'userid':'apple', 'name': '정윤서', 'hp':'010-1111-1111'}

# keys() : key 리스트를 리턴
print(dic2.keys())

# values() : value 리스트를 리턴
print(dic2.values())

# items() : key와 value를 한 쌍으로 묶는 튜플을 리턴
print(dic2.items())

# get() : key를 이용해서 value를 리턴
print(dic2.get('name'))
print(dic2.get("hp"))
print(dic2['name'])
print(dic2.get('age')) # 없는 키 입력시 None 리턴
print(dic2.get('age', '알수없음')) # 없는 키일 경우 None' 대신 '알수없음' 리턴, 있는 키일 경우 기존의 값 리턴


# in : key가 딕셔너리 안에 있는지 확인
print('name' in dic2) # True
print('age' in dic2) # False
```

### **1-5. 딕셔너리와 for**

```python
dic2 = {'no':1, 'userid':'apple', 'name': '정윤서', 'hp':'010-1111-1111'}

for i in dic2:
  print(i, end=' ') # key

print()

for i in dic2.keys():
  print(i, end=' ')

print()

for i in dic2.values():
  print(i, end=' ')

print()

for i in dic2:
  print(dic2[i], end=' ') # key값만큼 반복하며 value 얻기 - 인덱싱

print()

for i in dic2.keys():
  print(dic2.get(i)) # key값만큼 반복하며 value 얻기 - get 함수
```

### **문제.**
아래와 같은 딕셔너리 자료에서 주민등록번호를 추출하여 성별을 알아내는 프로그램을 작성해보자.  
(단, ssn의 7번째 자리의 성별은 1,3은 '남자', 2,4는 '여자'라고 한다)

> user = {"name":"김사과", "hp":"010-1111-1111", "ssn":"0010114068518"}


```python
# 내 풀이
user = {"name":"김사과", "hp":"010-1111-1111", "ssn":"0010114068518"}

jumin = user['ssn']
if jumin[6] == 1 or 3:
  print('남자')
if jumin[6] == 2 or 4:
  print('여자')
else:
  print('잘못된 주민등록번호입니다')
```

```python
# 모범답안
user = {"name":"김사과", "hp":"010-1111-1111", "ssn":"0010114068518"}

for i in user.keys():
  if i == 'ssn':
    ssn = user.get(i) # "0010114068518"
    if ssn[6] == '1' or ssn == '3':
      print('남자')
    elif ssn[6] == '2' or ssn[6] == '4':
      print('여자')
```
