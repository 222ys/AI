1. 스페셜 메서드(Special Method)

* __로 시작해서 __로 끝나는 특수함수
* 해당 메서드들을 구현하면 커스텀 객체에 여러가지 파이썬 내장 함수나 연산자를 적용 가능
* 오버라이딩 가능한 함수 목록은 아래 링크에서 확인  
(https://docs.python.org/3/reference/datamodel.html)

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def print_pt(self):
    print('({}, {})'.format(self.x, self.y)) # (3, 4)
  def __str__(self): # __str__ : 생성자에 기능을 오버라이딩
    return ('({}, {})'.format(self.x, self.y))
  def __add__(self, pt): # __add__ : + 기호에 기능을 오버라이딩
    new_x = self.x + pt.x
    new_y = self.y + pt.y
    return Point(new_x, new_y)
  def __sub__(self, pt): # __sub__ : - 기호에 기능을 오버라이딩
    new_x = self.x - pt.x
    new_y = self.y - pt.y
    return Point(new_x, new_y)
    return Point(new_x, new_y) # __mul__ : * 기호에 기능을 오버라이딩
  def __mul__(self, m):
    return Point(self.x * m , self.y * m) # __len__ : len 함수에 기능을 오버라이딩
  def __len__(self):
    return self.x ** 2 + self.y ** 2
  def __getitem__(self, index):
    if index == 0:
      return self.x
    elif index == 1:
      return self.y
    else:
      return -1
```

```python
p1 = Point(3, 4)
print(p1) # str이 생략된 형태
print(str(p1)) # 윗줄과 같은 식
p1.print_pt()
```

```python
p2 = Point(5, 8)
print(p2)

p3 = p1 + p2
print(p3)
```

```python
p4 = p1 - p2
print(p4)
```

```python
p5 = p1 * 3
print(p5)
```
