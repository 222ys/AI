# **1. 모듈**

* 모든 파이썬 파일(.py)은 모듈
* 클래스 또는 함수를 모아 놓은 파일

```
# 모듈 전체 가져오기
import 모듈명
# 사용
모듈명.함수명()

# 특정 함수만 골라서 가져오기
from 모듈명 import 함수1, 함수2..
from 모듈명 import *
# 사용
함수명()

# 모듈 별명 붙이기
import 모듈명 as 별명
# 사용
별명.함수명()

```

````python
import myutil # 모듈을 통째로 불러올 경우
````

````python
tot = myutil.sum(1, 10) # 모듈.함수() 로 사용
print(tot)
````

````python
from myutil import sum # 모듈에서 특정함수를 불러올 경우
````

````python
tot = sum(1, 10) # 함수()만 사용 가능
print(tot)
````

````python
import myutil as mu
````

````python
tot = mu.sum(1, 10)
print(tot)
````

# **2. 패키지**

### **2-1. 패키지란**
* 모듈의 상위개념
* 모듈이 모여있는 디렉토리

### **2-2. 패키지 관리자**
* 외부 모듈을 설치 또는 삭제하기 위한 프로그램
* pip 명령을 사용(파이썬 3.4 이후 파이썬을 설치하면 pip가 함께 설치됨)
* 윈도우 명령프롬프트 또는 터미널에서 사용
```
pip--help # pip 명령어 확인
pip install 패키지명 # 설치
pip uninstall 패키지명 # 삭제
```
