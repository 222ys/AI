# **1. 리스트(List)**

* 여러가지 데이터를 ,(콤마)로 구분하여 [ ]로 감싼 자료구조

```python
li1 = [1, 3, 5, 7]
print(li1)

li2 = list([1, 3, 5])
print(li2)

li3 = ['정윤서', '고수현']
print(li3)

li4 = [1, 1.5, '정윤서', True] # 타입이 다른 데이터를 포함, 개수도 미리 설정하지 않아도 됨
print(li4)

li5 = [1, 2, ['정윤서', '고수현']]
print(li5)
```

# **2. 리스트 인덱싱**

```python
li1 = [1, 3, 5, 7]
print(li1[0])
print(li1[1])
print(li1[-1])
print(li1[0]+li1[2])
```

```python
li2 = ['정윤서', '고수현', ['🎀', '🎗', '💘']]
print(li2[-1])
print(li2[-1][0]) # 2차원
```

```python
li3 = [1, 3, 5, ['정윤서', '고수현', ['❤', '💖']]]
print(li3[2])
print(li3[-2])
print(li3[-1][-2])
print(li3[-1][-1][0]) # 3차원
```

# **3. 리스트 슬라이싱**

```python
li4 = [10, 20, 30, 40, 50]
print(li4)
print(li4[0:3])
print(type(li4))

li5 = li4 # 주소를 공유하는 것
print(li5)
li5[0] = 100
print(li5)
print(li4) # 같은 주소를 가리키고 있기 때문에 같은 결과를 공유

li6 = [10, 20, 30, ['apple', 'banana', 'orange'], 40, ['❤', '💌']]
print(li6[2:6])
print(li6[5][:1])
```

# **4. 리스트 연산자**

```python
li7 = [10, 20, 30]
li8 = [40, 50, 60]

print(li7 + li8) # [10, 20, 30, 40, 50, 60]
print(li8 + li7) # [40, 50, 60, 10, 20, 30]
# print(li7 - li8) # TypeError: unsupported operand type(s) for -: 'list' and 'list'
print(li7 * 3) # [10, 20, 30, 10, 20, 30, 10, 20, 30]
print('❤' * 30)

li9 = [10, 20, 30]
li9[1] = '❤'
print(li9)
# print(li9[0] + '50') # TypeError: unsupported operand type(s) for +: 'int' and 'str'
print(li9[0] + li9[2])
print(li9[1] + '💦')
```

# **5. 리스트의 수정/삭제**

```python
li10 = [10, 20, 30]
li10[1] = 100
print(li10)

li10[1:2] = ['❤', '😍', '😎'] # 슬라이싱을 이용해 데이터를 변경할 경우 리스트에 데이터가 포함
print(li10)

li10 = [10, 20, 30]
li10[1] = ['❤', '😍', '😎'] # 인덱싱을 이용해 데이터를 변경할 경우 리스트 안에 리스트가 포함
print(li10)

print(li10[0:3])
li10[1:3] = [] # 슬라이싱으로 삭제
print(li10)

del li10[0] # del로 삭제
print(li10)
```

# **6. 리스트 함수**

```python
li11 = [10, 20, 30]

# append() : 리스트에 데이터를 추가
li11.append(100)
print(li11)

li11.append([40, 50])
print(li11)

# li11.append(60, 70) TypeError: append() takes exactly one argument (2 given)
# 요소 여러개를 넣고 싶다면 반복문에 append를 넣거나, 슬라이스로 넣으면 됨
```

```python
li12 = [10, 20, 30, 40, 50]

# pop() : 리스트의 마지막에 저장된 데이터를 반환하고 해당 데이터는 삭제
print(li12)
temp = li12.pop() # 데이터를 반환하여 변수에 저장
print(li12)
print(temp)
```

```python
li13 = [10, 20, 30, 10, 20, 80]

# remove() : 매개변수로 전달된 데이터를 제거. 처음 출현하는 데이터를 제거
li13.remove(20)
print(li13)

li13.remove(20)
print(li13)

#li13.remove(20) # 없는 데이터를 삭제 시도하면 에러
#print(li13)
```

```python
li14 = [10, 20, 30]

# insert() : 리스트의 특정 위치에 데이터를 삽입
li14.insert(1, 100) # 인덱스 1번 자리에 데이터 100을 삽입. 교체가 아니라 사이에 끼워주는 것임.
print(li14)
```

```python
li15 = [10, 20, 30, 40, 50]

# index() : 리스트에 저장된 데이터가 있을 경우 위치를 반환(없으면 에러)
print(li15.index(30))
# print(li15.index(130)) # ValueError: 130 is not in list
```

```python
li16 = ['😀', '😁', '😂', '🤣']

# reverse() : 리스트에 저장된 데이터를 반대의 순서로 변환
li16.reverse()
print(li16)
```

```python
li17 = [10, 40, 30, 100, 90, 50]

# sort() : 리스트에 저장된 데이터를 정렬. 단 같은 타입의 데이터끼리만 가능.
li17.sort() # 오름차순
print(li17)

li17.sort(reverse = True) # 내림차순
print(li17)

li18 = ['Apple', 'apple', 'orange', 'banana', 'melon']
li18.sort() # 아스키코드 순
print(li18)
```

```python
li19 = [10, 20, 30, 50, 20, 40, 30, 20]

# count() : 리스트에 포함된 특정 데이터의 개수를 반환
print(li19.count(20))
print(li19.count(15))
```

```python
li20 = [1, 2, 3]

# extend() : 리스트를 더함. append는 요소 하나 추가, extend는 요소 여러개 추가
li20.extend([40, 50])
print(li20)

li21 = [1, 2, 3]
li21 = li21 + [40, 50] # li21 += [40, 50]
```
