# **2. 출력 서식 지정하기**

1. %d(정수),%f(실수),%s(문자열)을 사용하여 출력 서식을 지정
2. 서식을 제외한 문자는 입력한 그대로 출력
3. 서식을 지정해 출력할 데이터가 2개 이상일 경우 ()로 데이터를 묶어야 함

```python
print('%d' % 100)
print('%d' % 100.12)

print('%f' % 100.12) # 소수점 여섯 자리까지 표현
print('%f' % 100)

print('%s' % 'Python')
print('%s' % 100)

print('%d + %d = %d' % (5, 3, 5+3))
```

```python
print('%5d' % 100)
print('%5d' % 10000)
print('%5d' % 100000) # 지정한 자릿수를 초과하는 경우 초과된 채 출력됨
print('%05d' % 100)
```

```python
print('%6.2f' % 123.45)
print('%6.2f' % 123.456) # 반올림
print('%6.2f' % 123.4) # 남은 자리는 0으로 채움
```

```python
print('%5s' % 'abc')
```

```python
# 5 / 3 = ?
print('%d / %d = %d' % (5, 3, 5/3)) # 결과를 %d로 출력하면 정수로 출력
print('%d / %d = %f' % (5, 3, 5/3)) # 결과를 %f로 출력하면 실수로 출력
```

# **3. format 함수와 출력 인덱스 사용**

```python
# 인덱스 : 0부터 순서대로 진행되는 정수
print('{0} {1}'.format('고수현', '정윤서'))
print('{1} {0}'.format('고수현', '정윤서'))
```

```python
print('{0:3s}님은 {1:03d}살입니다.'.format('정윤서', 23))
print('{0}님의 키는 {1:6.2f}cm입니다.'.format('정윤서', 160.5))
```
