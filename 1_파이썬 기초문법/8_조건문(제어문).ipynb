### **2-1-1. 비교연산자**
* <, >, >=, <=, !=, ==
* 결과가 True, False로 나옴

````python
print(10 > 5)
print(10 < 5)
print(10 >= 5)
print(10 <= 5)
print(10 != 5)
print(10 == 5)
````

### **2-2. if문**
```
if 조건식(결과가 True 또는 False가 나오는 문장):
  조건식이 True인 경우 실행할 문장
```

````python
age = int(input('나이를 입력하세요: '))

if age > 19:
  print('성인입니다.')
print('입력한 나이: ', age)
````

### **2-3. if ~ else문**

```
if 조건식(결과가 True 또는 False가 나오는 문장):
  조건식이 True인 경우 실행할 문장
  '''
else:
  조건식이 False인 경우 실행할 문장
  '''
```

````python
age = int(input('나이를 입력하세요: '))

if age > 19:
  print('성인입니다.')
  print('청불영화를 감상할 수 있습니다.')
else:
  print('미성년입니다.')
  print('디즈니를 보세요.')
print('입력한 나이: ', age)
````

### **2-4. if ~ elif ~ else문**
```
if 조건식(결과가 True 또는 False가 나오는 문장):
  조건식1이 True인 경우 실행할 문장
  '''
elif 조건식2:
  조건식2가 True인 경우 실행할 문장
  '''
elif 조건식3:
  조건식3이 True인 경우 실행할 문장
  '''
else:
  모든 조건식이 False인 경우 실행할 문장
  '''
```

````python
age = int(input('나이를 입력하세요: '))

if age > 19:
  print('성인입니다.')
elif age > 14:
  print('청소년입니다.')
elif age > 6:
  print('어린이입니다.')
elif age > 0:
  prit('유아입니다.')
else:
  print('입력이 잘못되었습니다.')
print('입력한 나이: ', age)
````

### **문제**
국어, 수학, 영어점수를 한번에 입력받아 총점, 평균, 학점을 구하는 프로그램을 만들어보자.
(단, 학점은 평균 기준이며 90점 이상은 A학점, 80점 이상은 B학점, 70점 이상은 C학점, 60점 이상은 D학점, 나머지는 F학점으로 함)

````python
#내 답안
kor, math, eng = map(int,input('국어, 수학, 영어 점수를 입력하세요.').split(' '))

tot = kor + math + eng
print('총점: ', tot)

avg = tot/3
print('평균: ', avg)

if avg >= 90:
  print('학점: ', 'A')
elif avg >= 80:
  print('학점: ', 'B')
elif avg >= 70:
  print('학점: ', 'C')
elif avg >= 60:
  print('학점: ', 'D')
else:
  print('학점: ', 'F')
````

````python
# 모범답안

kor, math, eng = map(int,input('국어, 수학, 영어 점수를 입력하세요.').split(' '))
tot = kor + math + eng
avg = tot/3

hak = '' #로컬변수로 보지 않도록 하기 위해 함수 밖에도 써주는 것이 좋음

if avg >= 90:
  hak = 'A'
elif avg >= 80:
  hak = 'B'
elif avg >= 70:
  hak = 'C'
elif avg >= 60:
  hak = 'D'
else:
  hak = 'F'

print('총점: ', tot)
print('평균: ', avg)
print('학점: ', hak)
````

### **2-5. AND 연산자**
* 양쪽 두 식의 값이 모두 True인 경우 결과가 True가 되는 연산자

> 정보처리기사의 합격 기준은 모든 과목의 평균 점수가 60점 이상이고, 각 과목의 점수가 40점 이상이다.

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|False|
|False|True|False|
|False|False|False|




````python
# 로그인 프로그램 작성
# 아이디 : admin, 비밀번호 : 1234

userid = input('아이디를 입력하세요')
userpw = input('비밀번호를 입력하세요')

if userid == 'admin' and userpw == '1234':
  print('로그인 되었습니다')
else:
  print('아이디 또는 비밀번호를 확인하세요')
````

### **문제**
주민등록번호를 입력받아 현재 나이를 구하는 프로그램을 작성해보자.

````python
juminNo = input('주민등록번호 13자리를 -없이 입력하세요') # 0010113068518

year = int(juminNo[:2]) # 001011 -> 00, 981011 -> 98

if int(juminNo[6]) <= 2: # 1900년대 태어났다면
  year = year +1900 # 98 + 1900 = 1998
else: # 2000대 태어났다면
  year = year + 2000 # 00 : 2000 = 2000

age = 2022 - year
print('현재 나이는 {}살입니다'.format(age))
````
