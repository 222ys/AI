# **1. 좌표 정렬하기**

https://www.acmicpc.net/problem/11650

2차원 평면 위의 점 N개가 주어진다. 좌표를 x좌표가 증가하는 순으로, x좌표가 같으면 y좌표가 증가하는 순서로 정렬한 다음 출력하는 프로그램을 작성하시오.

```python
n = int(input())
list = []

for _ in range(n): # _의 의미 : 반복할 때 굳이 _를 사용하지 않겠다는 의미
  x, y = map(int, input().split(' '))
  list.append((x, y))

list = sorted(list) # 알아서 다 정렬해줌

for i in list:
  print(i[0], i[1])
```

# **2. 소트인사이드**

https://www.acmicpc.net/problem/1427


배열을 정렬하는 것은 쉽다. 수가 주어지면, 그 수의 각 자리수를 내림차순으로 정렬해보자.

```python
# 내 답안
n = input() # n = '1234'
list = []
for i in range(len(n)): # range(4)
  list.append(int(n[i])) # list = [1, 2, 3, 4]
list.sort()
for i in range(len(n)):
  print(list[i], end = '')
```

```python
# 모범 답안
array = input()

for i in range(9, -1, -1):
  for j in array:
    if int(j) == i:
      print(i, end='')
```
