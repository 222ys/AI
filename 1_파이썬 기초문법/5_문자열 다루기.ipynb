# **2. 문자열 인덱싱**

```python
str = '외국대학에서 취득한 성적(학점)은 본교의 성적표에 그대로 표기하되, 전체 평균평점(GPA) 산정에는 포함되지 않습니다.'
print(str)
```

```python
print(str[0]) # 외
print(str[1]) # 국
print(str[20]) # 교
print(str[-1]) # . (마지막 문자)
print(str[-2]) # 다
```

# **3. 문자열 슬라이싱**

```python
str = '외국대학에서 취득한 성적(학점)은 본교의 성적표에 그대로 표기하되, 전체 평균평점(GPA) 산정에는 포함되지 않습니다.'
```

```python
print(str[0:8]) #인덱스 0부터 8이전까지 문자열
print(str[14:25])
print(str[22:]) #인덱스 22부터 끝까지 문자열
print(str[:21]) #인덱스 0부터 21이전까지 문자열
```

# **4. 문자열 함수**

```python
str = '외국대학에서 취득한 성적(학점)은 본교의 성적표에 그대로 표기하되, 전체 평균평점(GPA) 산정에는 포함되지 않습니다.'
```

```python
# len() : 인수로 입력된 문자열의 개수를 리턴, 파이썬 표준 함수
print(len(str))
```

```python
#count() : 문자열에서 인수로 입력된 문자열의 출현 횟수를 리턴, str 함수
print(str.count('성'))
print(str.count('GPA'))
print(str.count('gpa'))
```

```python
# find() : 인수로 입력된 문자열이 처음 나오는 위치를 인덱스로 리턴
print(str.find('성'))
print(str.find('a')) # 없으면 -1
```

```python
# index() : 인수로 입력된 문자열이 처음 나오는 위치를 인덱스로 리턴
print(str.index('성'))
print(str.index('a')) # 없으면 에러 ValueError: substring not found
```

```python
# replace() : 문자열을 치환
print(str.replace('외국대학', '해외교'))
```

```python
# upper() : 대문자로 변환
# lower() : 소문자로 변환
print(str.lower())
```

```python
str = '          Hello          Python          '
print(str)
```

```python
# lstrip() : 문자열 왼쪽 공백을 제거
print(str.lstrip())
```

```python
# rstrip() : 문자열 오른쪽 공백을 제거
print(str.rstrip())
```

```python
# strip() : 문자열 양쪽 공백을 제거
print(str.strip())
```

```python
# join() : 문자열에 특정 문자열을 삽입
str = 'Python'
print('🎀'.join(str))
```
