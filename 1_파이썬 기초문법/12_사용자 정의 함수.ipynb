### **1-1. 이름만 있는 함수**
```
def 함수명():
  함수가 호출되면 실행할 문장
  ...
```

```
함수명() # 호출
```

````python
def func1():
  print('처음으로 만드는 함수!')
````

````python
func1() # 호출을 해야 실행됨
func1()
````

````python
for i in range(5):
  func1()
````

### **1-2. 매개변수(parameter)가 있는 함수**
```
def 함수명(변수1, 변수2...):
  함수가 호출되면 실행할 문장
  ...
```

```
함수명(값1, 값2...) # 호출
```

````python
def func2(num):
  print('입력받은 숫자 : %d' % num)
````

````python
func2(10)
func2(30)
# func2(10, 20) # TypeError: func2() takes 1 positional argument but 2 were given
````

````python
def func3(start, end):
  sum = 0
  for i in range(start, end+1, 1):
    sum += i
  print('{}부터 {}까지의 합: {}'.format(start, end, sum))
````

````python
func3(1, 10)
func3(1, 100)
func3(10, 50)
````

### **1-5. 매개변수의 개수가 가변적인 함수**
```
def 함수명(*변수):
  함수가 호출되면 실행할 문장
  ...
  return 값
```

```
함수명(값1, 값2...) # 값을 몇 개 넣어도 상관없음
```

> 매개변수는 함수안에서 튜플형태로 사용

````python
def func7(*args):
  result = 0
  for i in args:
    result += i
  return result
````

````python
print(func7(10))
print(func7(10, 30, 20, 70, 50))
````

### **1-6. 매개변수를 딕셔너리 구조로 전달받는 함수**
```
def 함수명(**매개변수):
  함수가 호출되면 실행할 문장
  ...
  return 값
```

```
함수명(변수1=값1, 변수2=값2 ..) # 호출
```

````python
def func8(**args):
  return args

user = func8(userid='apple', name='정윤서', age=23)
print(user)
````

### **1-7. 리턴값이 여러개인 함수**
```
def 함수명(변수1, 변수2 ..):
  함수가 호출되면 실행할 문장
  ...
  return 값1, 값2
```

```
변수1, 변수2, .. = 함수명(값1, 값2, ..)
```

````python
def func9(num1, num2):
  return num1+num2, num1-num2, num1*num2, num1/num2
````

````python
result1, result12, result3, result4 = func9(10,3)
print(result1, result12, result3, result4)
````

### **문제.**
데이터 90, 80, 50, 100, 70을 매개변수로 전달해서 최대값, 최소값을 출력하는 함수를 만들어 활용해보자.  
(단, return을 사용할 것. 내장함수 sort를 사용하지 않음)

````python
# 내 풀이
def maxmin(*args):
  for i in range(len(args)-1):
    if args[i]<args[i+1]:
      max = args[i+1]
  for i in range(len(args)-1):
    if args[i]>args[i+1]:
      min = args[i+1]
  return max, min

max, min = maxmin(90, 80, 50, 100, 70)
print(max, min)
````

````python
# 모범 답안
def maxmin(*args): # (90, 80, 50, 100, 70)
  max, min = args[0], args[0]
  for i in args:
    if max < i:
      max = i

    if min > i:
      min = i
  return max, min

max, min = maxmin(90, 80, 50, 100, 70)
print('최대값 :', max)
print('최소값 :', min)
````
