````python
# ë‚´ í’€ì´
def maxmin(*args):
  for i in range(len(args)-1):
    if args[i]<args[i+1]:
      max = args[i+1]
  for i in range(len(args)-1):
    if args[i]>args[i+1]:
      min = args[i+1]
  return max, min

max, min = maxmin(90, 80, 50, 100, 70)
print(max, min)
````
